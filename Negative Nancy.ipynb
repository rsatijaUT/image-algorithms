# Fun code to replace an image by its negative

__Disclaimer: Not intended for political discourse__

```python
import numpy as np
```

```python
from PIL import Image
```

```python
from IPython.display import display
```

```python
im = Image.open('nancy.jpg')
```

```python
display(im)
```

```python
array=np.array(im)
print(array.shape)
```

```python
array
```

```python
# uint8 datatype means the integers are unsigned, i.e. non-negative, and max 8 bites per byte. The latter means 
# the max value of each integer in this array is 2^8-1=255
```

```python
mask = np.full(array.shape,255)
mask
```

```python
modified_array = mask-array

modified_array=modified_array.astype(dtype='uint8')

modified_array
```

```python
display(Image.fromarray(modified_array))
```
